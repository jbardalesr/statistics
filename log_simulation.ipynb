```python
import numpy as np
import matplotlib.pyplot as plt

color = ['red', 'green', 'black', 'blue']
# log(L(θ))
log_L = lambda theta, x_mean: x_mean*np.log(theta) + (1 - x_mean)*np.log(1 - theta)
# log(L(θ)) n -> inf
log_limit = lambda theta: p*np.log(theta) + (1 - p)*np.log(1 - theta)
# parametro a estimar
p = 1/3
# θ \in ]0, 1[
eps = 0.0002
theta_values = np.linspace(0 + eps, 1 - eps, 100)

for i, n in enumerate([5, 10, 50]):
    # muestra de tamaño n
    x_sample = np.random.choice([0, 1], size=n, p=[1-p, p])
    x_mean = x_sample.mean()
    # log(L(θ))
    log_values = log_L(theta_values, x_mean)
    # gráfico x_n
    plt.plot(theta_values, log_values, linestyle='dotted', label=f"n={n}",c=color[i])
    plt.axvline(x_mean, linestyle='dashed', c=color[i])
    
# gráfico en el límite
plt.plot(theta_values, log_limit(theta_values), linestyle='solid', label="limite")
plt.axvline(p, linestyle='dashed', c=color[-1])
plt.title(r"$\theta \rightarrow \overline{x_n}\log{(\theta)} + (1 - \overline{x_n})\log{(1 - \theta)}$")
plt.legend()
plt.show()
```
