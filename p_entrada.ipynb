##### 1. Dada una sucesión de v.a de Bernoulli. Utilice la ley de los grandes números y el TLC.

##### 2. Dado $X ∼ \mathcal N (3, 16)$. Determine el valor de $a$ en 

$$\mathbb P (X \le a) = 0.3$$

```python
q = normal.inv_cdf(0.3)
print("a =", q)
```

##### 3. Dado $X \sim \mathcal{N}(3,16)$. Determine los valores de $a$ y $b$ en
$$
\mathbb{P}(a \leq X \leq b)=0,3
$$
*Sugerencia: Utilice cuantiles y un lenguaje de programación.*


```python
alpha = 0.7
q1 = normal.inv_cdf(alpha / 2)
q2 = normal.inv_cdf(1 - alpha / 2)

print(f"a = {q1}, b = {q2}")
```

##### 4. Dada dos v.a. $X \sim \mathcal{N}\left(\theta_1, 1\right)$ y $Y \sim \mathcal{N}\left(\theta_2, 1\right)$ donde $\theta_1<\theta_2$. Indique que desigualdad es verdadera
$$
\mathbb{P}(X \leq t)<\mathbb{P}(Y \leq t), \quad \mathbb{P}(X \leq t)>\mathbb{P}(Y \leq t)
$$
*Sugerencia: Realice un gráfico.*


```python
plt.figure(figsize=(8, 3))

t = 2.5
theta1 = 1
theta2 = 3

x = np.linspace(-3, 7, 100)

dist1 = norm.pdf(x, loc=theta1, scale=1)
dist2 = norm.pdf(x, loc=theta2, scale=1)

plt.axvline(theta1, linestyle='dashed', c='r')
plt.axvline(theta2, linestyle='dashed', c='b')
plt.axvline(t, linestyle='dashed', c='k')

plt.plot(x, dist1, c='r')
plt.plot(x, dist2, c='b')
plt.show()
```
